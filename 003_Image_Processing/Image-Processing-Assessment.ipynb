```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
img = cv.imread("../DATA/giraffes.jpg")
show_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
display_img(show_img)
```

**TASK:Apply a binary threshold onto the image.**

```python
img_binary =cv.imread("../DATA/giraffes.jpg",0)

ret, thres = cv.threshold(img_binary,115,255,type = cv.THRESH_BINARY)
```

```python
display_img(thres, cmap = 'gray')
```

**TASK: Open the giaraffes.jpg file from the DATA folder and convert its colorspace to  HSV and display the image.**

```python
img =cv.imread("../DATA/giraffes.jpg")
img_hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)
```

```python
display_img(img_hsv)
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
kernel = np.ones((4,4), dtype = np.uint8)  / 10
```

```python
kernel
```

```python
output = cv.filter2D(show_img, -1, kernel) # (img, desired depth/ ddepth,kernel)
```

```python
display_img(output)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
img_grayscale = cv.imread("../DATA/giraffes.jpg",0)
```

```python
#calculating x gradient sobel
sobelx = cv.Sobel(img_grayscale, ddepth = cv.CV_64F, dx=1, dy=0, ksize= 5)
```

```python
display_img(sobelx, cmap='gray')
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
# plotting histogram

color = ["b","g","r"]

for i,col in enumerate(color):
    hist_values = cv.calcHist([show_img],channels=[i], mask = None, histSize=[256], ranges=[0,256])
    # plt.yscale('log')
    plt.plot(hist_values, color = col)
```
