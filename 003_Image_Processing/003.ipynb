## Image Thresholding
- In some Cv applications it is often necessary to convert color images to grayscale, since only edges and shape end up being important 
- Similarly, some applications only require a binary image showing general shapes

**Thresholding** is fundamentally a very simple method of segmenting an image into different parts.

- Thresholding will convert an image to consist of only two values, white or black.

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("../DATA/rainbow.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img) ;
```

```python
# Reading image as grayscale
img = cv.imread("../DATA/rainbow.jpg",0)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img) ;
```

```python
img.max()
```

```python
# Thresholding 

# thres is normally halfway i.e 255 / 2, maxval is 255 as possible highest value
ret, Thres1 = cv.threshold(img, thresh = 127 , maxval= 255, type = cv.THRESH_BINARY)
```

```python
ret
```

```python
# We have an binary image, all the values are 0 or 255
plt.imshow(Thres1) ;
```

```python
# Inversing binary image
ret2, Thres2 = cv.threshold(img, thresh = 127 , maxval= 255, type = cv.THRESH_BINARY_INV)
```

```python
ret2
```

```python
# Plotting image
plt.imshow(Thres2) ;
```

### Reading New Image

```python
img2 = cv.imread("../DATA/crossword.jpg",0)
```

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)
plt.imshow(img2, cmap = 'gray') ;
```

```python
# Creating function to " SHOW IMAGE"

def show_pic(imgage):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(imgage, cmap = "gray")
    
show_pic(img2)
```

### Simple Binary Threshold on Image


```python

ret_binary , threshold_binary = cv.threshold(img2,179,255, cv.THRESH_BINARY) 

# changing the first argument for threshold can be used to change the output
show_pic(threshold_binary)
```

### Adaptive Threshold on an Image

```python
thres_adaptive_1 = cv.adaptiveThreshold(img2,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY,11,8)
show_pic(thres_adaptive_1)
```

### Blending two threshold images


```python
blended = cv.addWeighted(src1 = threshold_binary, alpha = 0.5, src2= thres_adaptive_1, beta = 0.5, gamma = 0)
show_pic(blended)
```
