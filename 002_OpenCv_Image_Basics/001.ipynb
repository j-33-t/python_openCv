```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = cv.imread("../DATA/00-puppy.jpg")
```

```python
# Always check the type of variable to make sure you are reading a file
type(img) 
```

```python
img.shape
```

```python
plt.imshow(img) ;
```

- Matplotlib reading image order -> RGB
- Opencv reading image order -> BGR

hence the openCv image reading has swapped the order for image when we display using matplotlib

```python
# Converting BGR to RGB
img = cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img) ;
```

```python
plt.imshow(img[200:,650:1750]) ;
```

```python
img_gray = cv.imread("../DATA/00-puppy.jpg",cv.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
# Why output is green ? Because it is in viridis by default
plt.imshow(img_gray) ;
```

```python
img_gray = cv.cvtColor(img_gray,cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img_gray) ;
```

```python
plt.imshow(img_gray[200:,650:1750]) ;
```

```python
# Flipping Images [Upside Down]
new_img = cv.flip(img_gray,0)
plt.imshow(new_img) ;
```

```python
# Flipping Images [Mirror Image]
new_img = cv.flip(img_gray,1)
plt.imshow(new_img) ;
```

```python
# Flipping Images [Mirror Image + Upside Down]
new_img = cv.flip(img_gray,- 1)
plt.imshow(new_img) ;
```
